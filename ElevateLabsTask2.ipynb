```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



#load dataset
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
#mean
df.mean(numeric_only=True)
```

```python
#median
df.median(numeric_only=True)
```

```python
#mode
df.mode(numeric_only=True)
```

```python
#histogram
df.hist(figsize=(12,10))
plt.show()
```

```python
#boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["Age","Fare","SibSp","Parch"]])
plt.show()
```

```python
#corelation matrix
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(18,5))

# Survival Count
plt.subplot(1,3,1)
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')

# Gender Distribution
plt.subplot(1,3,2)
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')

# Passenger Class Distribution
plt.subplot(1,3,3)
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')

plt.tight_layout()
plt.show()
```

```python
### Observation
# - Most passengers did not survive.
# - The dataset contains more male passengers than female passengers.
# - Most passengers traveled in Third Class (Pclass = 3), followed by First and Second Class.
```
